```python
review_point_mean = reviews.points.mean()
def remean_points(row):
    row.points = row.points - review_point_mean
    return row

reviews.apply(remean_points, axis='columns')
```

```python
def change(row):
    if row.points >= 95 or row.country == 'Canada':
        row.points = 3
    elif row.points in range(85,95):
        row.points = 2
    else:
        row.points = 1
    return row    
```

```python
reviews = reviews.apply(change, axis='columns')
```

```python
reviews
```

```python
reviews.points.value_counts()
```

```python
reviews.points
```

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = reviews.apply(stars, axis='columns')
```

```python
star_ratings
```

```python
star_ratings.value_counts()
```
