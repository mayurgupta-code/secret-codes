```python
plt.scatter(x1,y1, c=colors, cmap='viridis')
plt.colorbar()
```

```python
#sizes of the dots

sizes = np.linspace(0,500,13)
```

```python
#alpha is for the transparency of the dots
plt.scatter(x1, y1, c=colors, cmap='nipy_spectral', s=sizes, alpha=0.5)
plt.colorbar()
```
