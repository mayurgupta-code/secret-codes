```python
import itertools

arr_1 = input().split()

arr = [int(x) for x in arr_1]

times = list(itertools.permutations(arr))

valid_times = []

for items in times:
    hour = str(items[0]) + str(items[1])
    hour = int(hour)
    
    if hour <= 23:
        minutes = str(items[2]) + str(items[3])
        minutes = int(minutes)
        
        total_minutes = hour*60 + minutes
        
        valid_times.append(total_minutes)
        

actual_min = max(valid_times)         

hour_ans = actual_min//60
min_ans = actual_min%60

print('{}:{}'.format(hour_ans,min_ans))
```

```python
arr_1 = input().split()

arr = [int(x) for x in arr_1]


while True:
    arr.sort()
    
    x,y = arr[-2:]
    
    if y == 1 and x == 0:
        print(y)
        break
    elif y == 0 and x == 0:
        print(y)
        break
    
    x_index = arr.index(x)
    y_index = arr.index(y)
    
    if x == y:
        arr.pop(x_index)
        arr.insert(x_index, 0)
        
        arr.pop(y_index)
        arr.insert(y_index, 0)
        
    else:
        arr.pop(x_index)
        arr.insert(x_index, 0)
        
        arr.pop(y_index)
        arr.insert(y_index, y-x)
        
        
```
