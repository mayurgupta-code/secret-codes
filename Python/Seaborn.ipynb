# Hello, Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('fifa.csv', index_col='Date', parse_dates=True)
```

```python
dataset.head()
```

```python
#set width and height of the figure
plt.figure(figsize=(20,6))

#line chart showing how fifa ranking has evolved over time
sns.lineplot(data=dataset)
```

```python
spotify_data = pd.read_csv('spotify.csv', index_col='Date', parse_dates=True)
```

```python
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
#line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
#set the width and heigth of the figure 
plt.figure(figsize=(20,10))

#add title
plt.title('Daily Global Streams of popular songs in 2017-2018')

#line chart
sns.lineplot(data=spotify_data)
```

### Plot a subset of the data

```python
#see the columns of the dataframes
spotify_data.columns
```

```python
plt.figure(figsize=(20,10))

plt.title('Daily Global Streams of Popular Songs in 2017-2018')

sns.lineplot(data=spotify_data['Shape of You'], label = 'Shape of You')
```

```python
plt.figure(figsize=(14,6))

plt.title('Daily Global Streams of Popular Songs in 2017-2018')

sns.lineplot(data=spotify_data['Shape of You'], label='Shape of You')
sns.lineplot(data=spotify_data['Despacito'], label='Despacito')

plt.xlabel('Date')
plt.ylabel('No. of Streams')
```

## Bar Chart

```python
plt.figure(figsize=(20,6))

sns.barplot(x=spotify_data.index, y=spotify_data['Despacito'])
```

```python
flight_data = pd.read_csv('flight_delays.csv', index_col='Month')
```

```python
flight_data
```

```python
sns.barplot(x=flight_data.index, y=flight_data['NK'])
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.title('Average arrival delay')

plt.ylabel('Arrival Delays (in minutes)')
```

## Heatmap

```python
plt.figure(figsize=(14,6))

sns.heatmap(data=flight_data, annot=True)

plt.title('Heatmap for the Delay Analysis')
plt.xlabel('Airline')
```

```python
flight_data.corr()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(data=flight_data.corr(), annot=True)
```

```python
plt.figure(figsize=(50, 20))

sns.heatmap(data=spotify_data)
```

## Scatter Plots

```python
insurance_data = pd.read_csv('insurance.csv')
```

```python
insurance_data
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
sns.scatterplot(x=insurance_data['charges'], y=insurance_data['bmi'])
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
#for categorical data scatter plot
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```
